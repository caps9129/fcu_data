# Scrap subject details 

```python
import pandas as pd
import requests
import json
import urllib3
from bs4 import BeautifulSoup
import curlify

s = requests.Session()

payload = {"baseOptions":{"lang":"cht",
                       "year":106,
                       "sms":2},
        "typeOptions":{"code":{"enabled":"false",
                               "value":""},
                       "weekPeriod":{"enabled":"true",
                                     "week":'',
                                     "period":"*"},
                       "course":{"enabled":"false",
                                 "value":""},
                       "teacher":{"enabled":"false",
                                  "value":""},
                       "useEnglish":{"enabled":"false"},
                       "specificSubject":{"enabled":"false",
                                          "value":"1"}
                      }
       }

# POST /Service/Search.asmx/GetType2Result HTTP/1.1
# Host: coursesearch03.fcu.edu.tw
# Connection: keep-alive
# Content-Length: 311
# Origin: https://coursesearch03.fcu.edu.tw
# User-Agent: Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.181 Safari/537.36
# Referer: https://coursesearch03.fcu.edu.tw/main.aspx?token=eyJ0eXAiOiJKV1QiLCJhbGciOiJIUzI1NiJ9.eyJleHAiOjE1MjUxOTYwNTV9.ZzdLb_OicST1W-wpNVO9P-GBq9xZlZl1T5eGdI0hNeM
# Accept-Encoding: gzip, deflate, br
# Accept-Language: en-US,en;q=0.9

headers = {
            'Accept': 'application/json, text/plain, */*',
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36',
            'Content-Type': 'application/json; charset=UTF-8',
 }

day_of_week = 1
class_schedule = pd.DataFrame
for day in range(1,8):
    
    payload['typeOptions']['weekPeriod']['week'] = str(day)

    res = s.post(url = 'https://coursesearch03.fcu.edu.tw/Service/Search.asmx/GetType2Result',
                 data=json.dumps(payload),
                 headers=headers)
    res.encoding = "utf-8"
    print(curlify.to_curl(res.request))
    print("====================================================")

    json_data = json.loads(res.content)
    json_data = json.loads(json_data['d'])

    if day == 1:
        class_schedule = pd.DataFrame(json_data['items'])
    else:
        class_schedule = class_schedule.append(pd.DataFrame(json_data['items']))
    
```

# Removing duplicates 

```python
class_schedule = class_schedule.drop_duplicates().reset_index(drop=True)
class_schedule
```

# Selecting relevant columns and removing duplicates

```python
sub_class_schedule = class_schedule[['scr_selcode','sub_name','scr_period']].copy()
sub_class_schedule = sub_class_schedule.drop_duplicates().reset_index(drop=True)
sub_class_schedule
```

# Remove classes with no venue

```python
sub_class_schedule = sub_class_schedule.drop(sub_class_schedule[sub_class_schedule['scr_period'].str.contains('未排教室')].index).reset_index(drop=True)
sub_class_schedule = sub_class_schedule.drop(sub_class_schedule[sub_class_schedule['scr_period'].str.split().str.len().lt(2)].index).reset_index(drop=True)
sub_class_schedule
```
